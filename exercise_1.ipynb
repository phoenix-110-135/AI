```python
# exercise_1: Array Creation and Manipulation
# تمرین_1: ایجاد و دستکاری آرایه
import numpy as np

print("1. Array Creation and Manipulation")
# Create a 3x3 array of random numbers between 1 and 100
#یک آرایه 3×3 از اعداد تصادفی بین 1 تا 100 ایجاد کنید
arr = np.random.randint(1, 101, size=(3, 3))
print("3x3 random array:")
print(arr)

# Find the minimum and maximum values
# حداقل و حداکثر مقادیر را پیدا کنید
print("Minimum amount:", np.min(arr))
print("Maximum value:", np.max(arr))

# Calculate the sum of each row
# مجموع هر ردیف را محاسبه کنید
print("Total of each row:")
print(np.sum(arr, axis=1))

# Calculate the product of each column
# حاصل ضرب هر ستون را محاسبه کنید
print("The product of each column:")
print(np.prod(arr, axis=0))

# Normalize the array
# آرایه را عادی کنید
arr_normalized = (arr - np.mean(arr)) / np.std(arr)
print("Normalized array:")
print(arr_normalized)
```

```python
# exercise_2 : Array Reshaping and Broadcasting
# تمرین_2: تغییر شکل آرایه و پخش
print("\n2. Array Reshaping and Broadcasting")
# Create a 1D array of numbers 1 to 12
# یک آرایه 1 بعدی از اعداد 1 تا 12 ایجاد کنید
arr1d = np.arange(1, 13)
print("1D array from 1 to 12:")
print(arr1d)

# Reshape to 4x3 array
# به آرایه 4x3 تغییر شکل دهید
arr3x4 = arr1d.reshape(3, 4)
print("3x4 array:")
print(arr3x4)

# Reshape to 4x3 array
# به آرایه 4x3 تغییر شکل دهید
arr4x3 = arr1d.reshape(4, 3)
print("4x3 array:")
print(arr4x3)

# Create a 3x1 array of ones and use spread to multiply with the 3x4 array
# یک آرایه 3x1 از یک ها ایجاد کنید و از spread برای ضرب در آرایه 3x4 استفاده کنید
arr3x1 = np.ones((3, 1))
arr3x4_broadcast = arr3x4 * arr3x1
print("3x4 array multiplied by 3x1 array:")
print(arr3x4_broadcast)

# Create a 1x4 array of doubles and use spread to add to the 3x4 array
#یک آرایه 1x4 از دوتایی ایجاد کنید و از spread برای اضافه کردن به آرایه 3x4 استفاده کنید
arr1x4 = np.full((1, 4), 2)
arr3x4_broadcast2 = arr3x4 + arr1x4
print("3x4 array with the addition of a 1x4 array:")
print(arr3x4_broadcast2)
```

```python
# exercise_3 : Indexing and Slicing
# تمرین_3 : نمایه سازی و برش
print("\n3. Indexing and Slicing")
# Create a 5x5 array of numbers 1 to 25
#یک آرایه 5×5 از اعداد 1 تا 25 ایجاد کنید
arr5x5 = np.arange(1, 26).reshape(5, 5)
print("آرایه 5x5 از 1 تا 25:")
print(arr5x5)

# Extract the central 3x3 subarray
# زیرآرایه مرکزی 3x3 را استخراج کنید
center_3x3 = arr5x5[1:4, 1:4]
print("زیرآرایه 3x3 مرکزی:")
print(center_3x3)

# Extract all elements in the second row
# تمام عناصر در ردیف دوم را استخراج کنید
row2 = arr5x5[1, :]
print("عناصر ردیف دوم:")
print(row2)

# Extract columns with odd index
# استخراج ستون ها با شاخص فرد
odd_cols = arr5x5[:, ::2]
print("ستون های با نمایه فرد:")
print(odd_cols)

# Extract elements larger than average with logical indexing
# عناصر بزرگتر از حد متوسط ​​را با نمایه سازی منطقی استخراج کنید
mean = np.mean(arr5x5)
greater_than_mean = arr5x5[arr5x5 > mean]
print("Elements larger than average:")
print(greater_than_mean)
```

```python
# exercise_4: Linear Algebra Operations
# تمرین_4: عملیات جبر خطی
print("\n4. Linear Algebra Operations")
# Create matrices A (2x3) and B (3x2)
# ایجاد ماتریس A (2x3) و B (3x2)
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[7, 8], [9, 10], [11, 12]])

# Calculate the matrix product A.B
# محاسبه حاصل ضرب ماتریس A.B
AB = np.matmul(A, B)
print("Calculate the matrix product A.B :")
print(AB)

# Calculate the matrix product B.A
# محاسبه حاصل ضرب ماتریس B.A
BA = np.matmul(B, A)
print("Calculate the matrix product B.A :")
print(BA)

# Find the eigenvalues ​​and eigenvectors of A.B
# مقادیر ویژه و بردارهای ویژه A.B را پیدا کنید
eigenvalues, eigenvectors = np.linalg.eig(np.matmul(A, B))
print("Eigenvalues ​​and eigenvectors of A.B :")
print("eigenvalues :", eigenvalues)
print("eigenvectors :", eigenvectors)

# determinant B.A
det_BA = np.linalg.det(np.matmul(B, A))
print("determinant B.A:", det_BA)
```
